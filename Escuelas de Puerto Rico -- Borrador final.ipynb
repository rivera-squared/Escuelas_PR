# Proyecto Escuelas de Puerto Rico

##  Trasfondo

Bajo la administración Rosselló hijo se cerraron mas de dos centenares de escuelas . Funcionarios del gobierno citaron un reajuste necesario para ser mas efectivos y así lograr un ahorro para el Departamento de Educación. Es de conocimiento común la situación precaria que viven todos en la isla y que es necesario “bregar con lo que hay”. Sin embargo, ¿Qué datos se utilizaron para determinar cuales planteles cerraban y cuales permanecerían abiertas? En mi opinión, no se hizo la pregunta más importante: “¿Cómo evaluarían los funcionarios públicos el rendimiento de los estudiantes luego del cierre de las escuelas? Dado el hecho ninguna de las administraciones Rosselló, Vázquez ni Pierluisi ha arrojado luz sobre el rendimiento de los estudiantes luego del cierre de escuelas, me he dado la tarea de estudiar un poco mas a fondo la interrogante.

## Proceso de obtener y crear una base de datos

Para obtener los datos pertinentes, primero me contacté en octubre del 2020 con el Departamento de Educación vía email, donde me identifiqué como estudiante graduado de universidad, para peticionarle los datos disponibles sobre las escuelas públicas. Desafortunadamente, mis intentos de conseguir algún funcionario no fueron fructíferos así que procedí ha derivar mi propia base de datos utilizando la información disponible en el dashboard para cada una de las escuelas del sistema público puertorriqueño. Primero, visité el enlace (https://schoolreportcard.azurewebsites.net/reportes/perfilescolar) para obtener la lista de cada una de las escuelas por región para los años escolares 2016-17 y 2017-18. Del dashboard conseguí la información de la matrícula, el nivel de la escuela y la proficiencia de las asignaturas español, matemáticas, inglés y ciencia. 

Cabe destacar que el puntaje que se le asigna a las escuelas bajo proficiencia significa el porcentaje de estudiantes por escuela que obtuvieron al menos 80% en las pruebas META-PR. En otras palabras, si bajo la proficiencia en español una escuela fue asignada 50, quiere decir que el 50% de los estudiantes que tomaron la prueba obtuvieron al menos 80% en la META-PR de español ese año escolar. Luego de haber completado la lista de todas las escuelas del archipiélago, visité el enlace (https://schoolreportcardstorage.z13.web.core.windows.net/index.html) para verificar que toda la información estuviera correcta y también para conseguir los nombres de las escuelas que habían cerrado como parte de la política pública. 

Luego de haber verificado que toda la información estuviera correcta, busqué la lista publicada por el Departamento de Educación que indicaba que escuelas que habían sido seleccionadas para recibir estudiantes de planteles que cerraron. De esa manera, creé una variable dicótoma nueva, “escuela_receptora”, que indica si la escuela recibió estudiantes de planteles cerrados—la idea detrás de escuela_receptora es que sería utilizada como variable de control cuando usara modelos de regresión (mas en secciones subsiguientes).

Al finalizar el proceso, la base de datos culminó con 1,093 entradas (escuelas) cubriendo tres años escolares (2016-17, 2017-18 y 2018-19). *Al momento de este escrito solo cuento con tres años escolares; sin embargo, estaré actualizando la base de datos para que así refleje la información más reciente.*

## Fase de exploración de datos

Base de datos en mano, podemos comenzar a hacer algunas preguntas iniciales. La primera interrogante que surgio fue, ¿Cuantos planteles cerraron por municipio? 

En el mapa presentado arriba, entre mas oscuro sea el color, el mayor número de escuelas que se cerraron por municipio. El color gris denota que ninguna escuela se cerró en dicho municipio.

San Juan, como puede observarse, fue el municipio que más escuelas cerró (22 planteles) en términos absolutos. Por su tamaño y por ser ciudad capital, uno puede inferir que San Juan tiene mas escuelas que los demás municipios. Para entender el problema mas profundamente, decidí graficar las escuelas cerradas por municipio pero, esta vez, en términos porcentuales. 

Como podemos apreciar en esta nueva gráfica, en términos porcentuales, Aguadilla y Sabana Grande perdieron 50% de sus planteles, seguidos por Salinas y Dorado (47.1% y 45.5%, respectivamente), luego de la implementación de cierre de escuelas.

Ya sabemos qué municipios se vieron mas afectados luego del cierre de escuelas pero, ¿acaso los datos demuestran que el cierre de escuelas afectó a estudiantes?

## Hipótesis

Mi primera hipótesis sobre el cierre de escuelas, como es de esperarse, es que afectaría a los estudiantes de Puerto Rico. Ya teniendo la base de datos de todos los planteles del archipiélago, y habiendo identificado las escuelas que fueron seleccionadas para recibir estudiantes de planteles que cerraron, tengo un buen punto de partida para hacer mis primeras comparaciones entre los tipos de planteles.

Para comenzar, usaré los niveles de proficiencia por cada asignatura comparar los desempeños de los diferentes tipos de escuelas: los que recibieron estudiantes de planteles cerrados y las escuelas que no recibieron estudiantes. Siguiendo mi hipótesis, estudiantes que cursan en escuelas que recibieron estudiantes de planteles cerrados deberían, en promedio, de tener niveles de proficiencia (en cada una de las disciplinas) mas bajos que sus compañeros que cursan en escuelas que no recibieron estudiantes de planteles cerrados.

## Método

Debo enfatizar que este estudio se enfoca en el desempeño académico de las escuelas como tal, y de ahí tomar el resultado para inferir su efecto en los estudiantes. La razón para este método es debido a la naturaleza de los datos disponibles desde el Departamento de Educación. Para un entendimiento más profundo y exhaustivo del efecto del cierre de escuelas de Puerto Rico en sus estudiantes, datos a nivel estudiante son necesarios. A pesar de los contratiempos existentes para un minuicioso estudio, un enfoque macro es un buen inicio para comenzar nuestro proyecto.

## Segundo Análisis: Modelos de Regresión Linear

Como el comparar los histogramas no son suficiente para concluir que las escuelas que recibieron estudiantes de planteles cerrados, por lo general, les va peor que sus contrapartes, acudo al Modelo de Regresión Linear.

## Resumen de coeficientes: Modelo Linear


Para comenzar el análisis de regresión linear, decidí hacer dos modelos diferentes. Modelo #1 tiene como variable dependiente el nivel de proficiencia para cada una de las asignaturas de META-PR (español, matemática, inglés y ciencias) y su variable independiente la variable dicótoma ‘escuela_receptora’ (1 = escuela recibió estudiante de planteles cerrados; 0 = escuela no recibió estudiantes de planteles cerrados). Para el Modelo # 2, añadí la variable ‘estudiantes_maestro’ que reseña el promedio de estudiantes por maestro por cada escuela.

Con varias excepciones, todos los modelos enfatizan la hipótesis que, en promedio, niños que cursan en escuelas que recibieron estudiantes de planteles cerrados les va peor que sus contrapartes en escuelas donde no recibieron estudiantes.

Antes de describir mas a fondo cada uno de los hallazgos, quisiera enfatizar que los modelos explican muy poco (medido por el bajo nivel de R2) el efecto que tiene el recibir estudiantes de planteles cerrados en el rendimiento académico. Esto puede ser debido a que utilizo pocas variables pertinentes para el estudio (expliqué mis limitaciones en una sección anterior). Sin embargo, se puede notar que, cuando se añaden mas variables pertinentes (como por ejemplo la adición de ‘estudiantes_maestro’ en el modelo #2), el coeficiente cambia de una manera consistente con la hipótesis; además que R2 mejora. Es mi esperanza el conseguir información de mas variables relevantes (i.e. cantidad promedio invertida por estudiante por escuela, cantidad de computadoras por escuela, ingreso promedio de los padres por escuela, etc…) para poder mejorar los modelos aquí creados.

Para la META-PR Español, el modelo de regresión linear estimó los coeficientes de la variable ‘escuela_receptora’ como -3.78 y -4.19 (Modelo #1 y Modelo #2, respectivamente). En términos simples, el modelo estimó que, en promedio, si una escuela recibió estudiantes de planteles cerrados, el nivel de proficiencia para META-PR Español sería 3.78 o 4.19 (dependiendo el modelo) puntos mas bajo comparado con el nivel de proficiencia para los planteles que no recibieron estudiantes. En ambos modelos, el coeficiente ‘escuela_receptora’ es estadísticamente significante (H0 = 0 es rechazada; p < 0.005). Finalmente, el modelo #2 estimó el coeficiente de la variable estudiantes_maestro como -0.8; por cada punto que incremente la proporción de estudiantes por maestro por escuela, el nivel de proficiencia disminuye por .8 puntos. Al igual que la variable de ‘escuela_receptora’, ‘estudiantes_maestro’ es estadísticamente significante.

Para la META-PR Matemática, el modelo de regresión linear estimó los coeficientes de la variable ‘escuela_receptora’ como 3 y 2.17 (Modelo #1 y Modelo #2, respectivamente). Contrario a las demás asignaturas, los coeficientes de ambos modelos sugieren que la proficiencia en matemáticas en las escuelas que recibieron estudiantes de planteles cerrados, en promedio, son mejores que en las escuelas que no recibieron estudiantes. Esto suena contra intuitivo a la hipótesis original de que las escuelas que reciben estudiantes, en promedio, les va peor. Sin embargo, tengo que señalar varias observaciones. El coeficiente estimado de ‘escuela_receptora’ en el modelo #2 (cuando se le añade una variante relevante mejorando el estimado), disminuye, lo cual es indicativo que el estimado continuaría en la misma dirección que los demás coeficientes en las diferentes asignaturas. Otra observación que debo señalar es que la hipótesis nula de H0 = 0 no puede ser rechazada (p > .10); en otras palabras, la variable ‘escuela_receptora’ no es estadísticamente significante. La otra observación que es importante resaltar es que la variable ‘estudiantes_maestro’ en el modelo #2 es estadísticamente significante y, igual que la misma variable en los demás modelos, también es negativa. Al ser el estimado negativo, apunta a nuestras sospechas que entre mas la proporción de estudiantes por maestro sea, mas se verá afectado el desempeño académico.

Para la META-PR Inglés, el modelo de regresión linear estimó los coeficientes de la variable ‘escuela_receptora’ como -4.38 y -4.74 (Modelo #1 y Modelo #2, respectivamente). En términos simples, el modelo estimó que, en promedio, si una escuela recibió estudiantes de planteles cerrados, el nivel de proficiencia para META-PR Inglés sería 4.38 o 4.74 (dependiendo el modelo) puntos mas bajo comparado con el nivel de proficiencia para los planteles que no recibieron estudiantes. En ambos modelos, los coeficientes de ‘escuela_receptora’ son estadísticamente significantes. Finalmente, el modelo #2 estimó el coeficiente de la variable ‘estudiantes_maestro’ como -0.7; por cada punto que incremente la proporción de estudiantes por maestro por escuela, el nivel de proficiencia disminuye por .7 puntos. Al igual que la variable de ‘escuela_receptora’, ‘estudiantes_maestro’ es estadísticamente significante.

Para la META-PR Ciencias, el modelo de regresión linear estimó los coeficientes de la variable ‘escuela_receptora’ como -2.21 y -2.82 (Modelo #1 y Modelo #2, respectivamente). Para esta asignatura, el coeficiente de ‘escuela_receptora no es estadísticamente significante (hipótesis nula de H0 = 0 no puede ser rechazada; p > .10). Sin embargo, cuando observamos el mismo coeficiente para el modelo #2, nos damos cuenta de que disminuye y sí es estadísticamente significante (aunque esta vez α = .10; p < .10). Esto, al igual que para los coeficientes ‘escuela_receptora’ en ambos modelos de META-PR Matemática, es indicativo de que el efecto de recibir estudiantes de escuelas cerradas es nocivo para el rendimiento académico de sus estudiantes. El modelo #2 estimó el coeficiente de la variable ‘estudiantes_maestro’ como -1.16; o por cada punto que incremente la proporción de estudiantes de estudiantes por maestro por escuela, el nivel de proficiencia disminuye por 1.16 puntos. Finalmente, la variable es 
estadísticamente significante.

##  Conclusiones y próximos pasos

Como pudimos apreciar, las comparaciones de los histogramas y los resultados del análisis de regresión linear nos muestran que, de hecho, estudiantes en escuelas donde recibieron pupilos de planteles cerrados se vieron afectados. Utilicé la métrica del nivel de proficiencia por asignatura como lo estableció el Departamento de Educación en su dashboard. Este nivel de proficiencia, recordemos, es el porcentaje de estudiantes por escuela que obtuvieron al menos 80% en la META-PR por asignatura. La idea de esta métrica en los modelos de regresión es obtener la diferencia de rendimientos en los diferentes tipos de escuelas y de ahí inferir el resultado en los estudiantes. Estos análisis no resaltan, lamentablemente, las variables que afectan a los estudiantes en la realidad; sin embargo, este análisis, aunque rudimentario, es un excelente inicio para entender mas objetivamente las consecuencias de la implementación del cierre de escuelas en Puerto Rico. Es mi mas sincero deseo el poder conseguir mas datos sobre las escuelas y los (las) estudiantes puertorriqueños (as) y continuar analizando las dinámicas de las políticas públicas.
