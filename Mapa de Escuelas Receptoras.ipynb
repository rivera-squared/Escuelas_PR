# ¿Dónde están ubicadas las escuelas receptoras en Puerto Rico?

```python
import pandas as pd
import numpy as np
import os
```

```python
# Import the folium library
import folium
```

```python
os.chdir('/Users/edwardriverarivera/Documents/Python/educacion/')
```

```python
escuelas = pd.read_csv('escuelas.csv') # Cargando el df

escuelas = escuelas.loc[:, ['longitude','latitude','escuela_receptora','escuela']] # Seleccionando las columnas de interés
receptoras = escuelas[escuelas['escuela_receptora'] == 1] #Filtrando por las escuelas receptoras
```

```python
# Coordenadas del mapa de PR
map = folium.Map(location=[18.200178, -66.064513])
```

```python
# Añadiendo las coordenadas de las escuelas en el mapa
for i in range(0,len(receptoras)):
   folium.Marker(
      location=[receptoras.iloc[i]['latitude'], receptoras.iloc[i]['longitude']],
      popup=receptoras.iloc[i]['escuela'],
   ).add_to(map)
```
