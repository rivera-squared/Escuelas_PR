# Proyecto Escuelas de Puerto Rico

##  Trasfondo

Bajo la administración Rosselló hijo se cerraron mas de dos centenares de escuelas . Funcionarios del gobierno citaron un reajuste necesario para ser mas efectivos y así lograr un ahorro para el Departamento de Educación. Es de conocimiento común la situación precaria que viven todos en la isla y que es necesario “bregar con lo que hay”. Sin embargo, ¿Qué datos se utilizaron para determinar cuales planteles cerraban y cuales permanecerían abiertas? En mi opinión, no se hizo la pregunta más importante: “¿Cómo evaluarían los funcionarios públicos el rendimiento de los estudiantes luego del cierre de las escuelas? Dado el hecho ninguna de las administraciones Rosselló, Vázquez ni Pierluisi ha arrojado luz sobre el rendimiento de los estudiantes luego del cierre de escuelas, me he dado la tarea de estudiar un poco mas a fondo la interrogante.

## Proceso de obtener y crear una base de datos

Para obtener los datos pertinentes, primero me contacté en octubre del 2020 con el Departamento de Educación vía email, donde me identifiqué como estudiante graduado de universidad, para peticionarle los datos disponibles sobre las escuelas públicas. Desafortunadamente, mis intentos de conseguir algún funcionario no fueron fructíferos así que procedí ha derivar mi propia base de datos utilizando la información disponible en el dashboard para cada una de las escuelas del sistema público puertorriqueño. Primero, visité el enlace (https://schoolreportcard.azurewebsites.net/reportes/perfilescolar) para obtener la lista de cada una de las escuelas por región para los años escolares 2016-17 y 2017-18. Del dashboard conseguí la información de la matrícula, el nivel de la escuela y la proficiencia de las asignaturas español, matemáticas, inglés y ciencia. 

Cabe destacar que el puntaje que se le asigna a las escuelas bajo proficiencia significa el porcentaje de estudiantes por escuela que obtuvieron al menos 80% en las pruebas META-PR. En otras palabras, si bajo la proficiencia en español una escuela fue asignada 50, quiere decir que el 50% de los estudiantes que tomaron la prueba obtuvieron al menos 80% en la META-PR de español ese año escolar. Luego de haber completado la lista de todas las escuelas del archipiélago, visité el enlace (https://schoolreportcardstorage.z13.web.core.windows.net/index.html) para verificar que toda la información estuviera correcta y también para conseguir los nombres de las escuelas que habían cerrado como parte de la política pública. 

Luego de haber verificado que toda la información estuviera correcta, busqué la lista publicada por el Departamento de Educación que indicaba que escuelas que habían sido seleccionadas para recibir estudiantes de planteles que cerraron. De esa manera, creé una variable dicótoma nueva, “escuela_receptora”, que indica si la escuela recibió estudiantes de planteles cerrados—la idea detrás de escuela_receptora es que sería utilizada como variable de control cuando usara modelos de regresión (mas en secciones subsiguientes).

Al finalizar el proceso, la base de datos culminó con 1,093 entradas (escuelas) cubriendo tres años escolares (2016-17, 2017-18 y 2018-19). *Al momento de este escrito solo cuento con tres años escolares; sin embargo, estaré actualizando la base de datos para que así refleje la información más reciente.*

## Fase de exploración de datos

Base de datos en mano, podemos comenzar a hacer algunas preguntas iniciales. La primera interrogante que surgio fue, ¿Cuantos planteles cerraron por municipio? 

En el mapa presentado arriba, entre mas oscuro sea el color, el mayor número de escuelas que se cerraron por municipio. El color gris denota que ninguna escuela se cerró en dicho municipio.

San Juan, como puede observarse, fue el municipio que más escuelas cerró (22 planteles) en términos absolutos. Por su tamaño y por ser ciudad capital, uno puede inferir que San Juan tiene mas escuelas que los demás municipios. Para entender el problema mas profundamente, decidí graficar las escuelas cerradas por municipio pero, esta vez, en términos porcentuales. 

Como podemos apreciar en esta nueva gráfica, en términos porcentuales, Aguadilla y Sabana Grande perdieron 50% de sus planteles, seguidos por Salinas y Dorado (47.1% y 45.5%, respectivamente), luego de la implementación de cierre de escuelas.

Ya sabemos qué municipios se vieron mas afectados luego del cierre de escuelas pero, ¿acaso los datos demuestran que el cierre de escuelas afectó a estudiantes?

## Hipótesis

Mi primera hipótesis sobre el cierre de escuelas, como es de esperarse, es que afectaría a los estudiantes de Puerto Rico. Ya teniendo la base de datos de todos los planteles del archipiélago, y habiendo identificado las escuelas que fueron seleccionadas para recibir estudiantes de planteles que cerraron, tengo un buen punto de partida para hacer mis primeras comparaciones entre los tipos de planteles.

Para comenzar, usaré los niveles de proficiencia por cada asignatura comparar los desempeños de los diferentes tipos de escuelas: los que recibieron estudiantes de planteles cerrados y las escuelas que no recibieron estudiantes. Siguiendo mi hipótesis, estudiantes que cursan en escuelas que recibieron estudiantes de planteles cerrados deberían, en promedio, de tener niveles de proficiencia (en cada una de las disciplinas) mas bajos que sus compañeros que cursan en escuelas que no recibieron estudiantes de planteles cerrados.

## Método

Debo enfatizar que este estudio se enfoca en el desempeño académico de las escuelas como tal, y de ahí tomar el resultado para inferir su efecto en los estudiantes. La razón para este método es debido a la naturaleza de los datos disponibles desde el Departamento de Educación. Para un entendimiento más profundo y exhaustivo del efecto del cierre de escuelas de Puerto Rico en sus estudiantes, datos a nivel estudiante son necesarios. A pesar de los contratiempos existentes para un minuicioso estudio, un enfoque macro es un buen inicio para comenzar nuestro proyecto.
