# El pastel que es el Departamento de Educación

```python
# Importing necessary components
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import squarify 
import plotly.express as px
```

```python
df = pd.read_excel('presupuesto_DE_resumen.xlsx')
```

```python
df2022 = df[df['Ano Fiscal'] == 2022]
df2022 = df2022.drop(columns = 'Total')
```

```python
variables = ['Nomina y Costos Relacionados',
       'Facilidades y Pagos por Servicios Publicos', 'Servicios Comprados',
       'Donativos, Subsidios y Otras Distribuciones',
       'Gastos de Transportacion y Subsistencia', 'Servicios Profesionales',
       'Otros Gastos', 'Materiales y Suministros', 'Compra de Equipo',
       'Anuncios y Pautas en Medios',
       'Incentivos y Subsidios Dirigidos al Bienestas de la Ciudadania',
       'Aportaciones a Entidades no Gubernamentales',
       'Asignacion Pareo de Fondos Federales', 'Asignaciones Englobadas']
```

```python
x=pd.melt(df2022,
        id_vars=["Ano Fiscal", "Dpto"],
        value_vars= variables,
        value_name="Gasto")
```

```python
X = x[x['Gasto'] > 0]
X = X[X['Dpto'] == "Escuelas de la Comunidad"]
```

# ¿Cómo se divide el pastel de las escuelas de PR?

```python
fig = px.treemap(X, 
                 path=['Dpto', 'variable'], 
                 values='Gasto',
                 color='Gasto'
                )
fig.show()
```

```python
y = x[x['Gasto'] > 0]
```
